Si deve studiare il dataset delle vendite di amazon nell'anno 2025.
Il dataset è parziale, non è completo di tutte le vendite effettuate nell'anno 2025.

Per prima cosa si vuole mostrare com'è composto il dataset, in modo da effettuare un'analisi esplorativa, ovvero un'analisi descrittiva e rappresentando i dati in vari grafici, per capirne meglio il legame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

dt = pd.read_csv(f'C:\\Users\\Media\\Desktop\\LG_LabProgramming_II\\ProgrammingLab_II\\Esercizi_Laboratorio\\99_Esercitazioni\\amazon_sales_data 2025.csv', encoding='latin1')

display(dt.head(3))
dt.columns

# Voglio verificare se ci sono vendite uguali e chi li ha comprati
#GroupBy permette di raggruppare i dati in base alla colonna 'Product'
# e sommare le quantità vendute per ciascun prodotto
# La funzione 'sum' viene applicata alla colonna 'Quantity' per ottenere 
# il totale delle vendite per ciascun prodotto
# 'reset_index' viene utilizzata per ripristinare l'indice del DataFrame risultante
# e creare un nuovo DataFrame con i risultati
prodotti_venduti = dt.groupby('Product')['Quantity'].sum().reset_index()

#Riordino i prodotti in base alla quantità venduta in ordine decrescente
prodotti_venduti = prodotti_venduti.sort_values(by='Quantity', ascending=False)

#voglio creare un grafico a barre per visualizzare i prodotti piu venduti
plt.figure(figsize=(15,6)) #Definisco la dimensione della figura
sns.barplot(x='Product', y='Quantity', data=prodotti_venduti, palette='viridis') #Uso Seaborn per creare un grafico a barre
plt.title('Boxplot dei prodotti più venduti')
plt.xlabel('Prodotto')
plt.ylabel('Quantità venduta')
plt.xticks(rotation=90) #Ruoto le etichette dell'asse x di 90 gradi per una migliore leggibilità
plt.grid(axis='y') #Aggiungo una griglia solo sull'asse y
plt.show() #Mostro il grafico

#Ora voglio calcolare il totale delle vendire per ciascun prodotto
#ed il suo prezzo complessivo, costo per prodotto
#Soldi spesi dagli utenti per acquistare i prodotti
prezzo_prodotto = dt.groupby('Product')['Price'].sum().reset_index() #Permette di raggruppare i deti in base alla colonna prodotto e contare il costo di ogni prodotto venduto

#Creo il grafico a barre
plt.figure(figsize=(15,6)) #Definisco la dimensione della figura
sns.catplot(x='Product', y='Price', data=prezzo_prodotto, kind='bar', palette='viridis') #Uso Seaborn per creare un grafico a barre
plt.title('Boxplot dei prodotti più venduti')
plt.xlabel('Prodotto')
plt.ylabel('Prezzo totale')
plt.xticks(rotation=90) #Ruoto le etichette dell'asse x di 90 gradi per una migliore leggibilità
plt.grid(axis='y') #Aggiungo una griglia solo sull'asse y
plt.show() #Mostro il grafico


#Realizzare un'analisi temporale delle vendite, oppure vedere in quale periodo
#Sono stati effettuati più acquisti
#Oppure il periodo con incassi maggiori rapportati alle vendite




```
